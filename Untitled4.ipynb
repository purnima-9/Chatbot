```python
import pandas as pd
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import GOOG  # Replace this with your custom data

# Function to calculate the RSI
def compute_rsi(data, period=14):
    # Convert data to pandas Series to perform rolling calculations
    data_series = pd.Series(data)
    price_diff = data_series.diff()
    gains = price_diff.where(price_diff > 0, 0).rolling(window=period).mean()
    losses = -price_diff.where(price_diff < 0, 0).rolling(window=period).mean()
    relative_strength = gains / losses
    return 100 - (100 / (1 + relative_strength))

# Custom trading strategy class
class EMAVWAPTradingStrategy(Strategy):
    # Define trading parameters
    stop_loss_percentage = 0.025  # 2.5%
    take_profit_percentage = 0.60  # 60%
    atr_length = 14  # ATR period for volatility-based adjustments
    rsi_length = 14  # RSI period for market filtering

    def init(self):
        # Calculate RSI using the defined function and store in self.rsi
        self.rsi = self.I(lambda x: compute_rsi(x, self.rsi_length).to_numpy(), self.data.Close)
        # Track position status (0: no position, 1: long, -1: short)
        self.position_status = 0

    def next(self):
        # Get the latest closing price and RSI value
        current_price = self.data.Close[-1]
        current_rsi = self.rsi[-1]

        # Logic for entering or exiting positions based on RSI
        if self.position_status == 0:
            # Long entry if RSI is above 60
            if current_rsi > 60:
                self.buy()
                self.position_status = 1

            # Short entry if RSI is below 40
            elif current_rsi < 40:
                self.sell()
                self.position_status = -1

        elif self.position_status == 1:
            # Close long position if RSI falls below 40
            if current_rsi < 40:
                self.position.close()
                self.sell()  # Sell to exit long
                self.position_status = -1

        elif self.position_status == -1:
            # Close short position if RSI rises above 60
            if current_rsi > 60:
                self.position.close()
                self.buy()  # Buy to exit short
                self.position_status = 1

        # Risk management: Close position at the end of the day to prevent overnight exposure
        if self.data.index[-1].time() >= pd.Timestamp("15:30").time():
            if self.position_status != 0:
                self.position.close()
                self.position_status = 0

# Example usage of the strategy with stock data (use real intraday data for a live test)
backtest = Backtest(GOOG, EMAVWAPTradingStrategy, cash=10000, commission=0.002, trade_on_close=True)

# Run the backtest
results = backtest.run()
print(results)

# Plot the backtest results
backtest.plot()
```
